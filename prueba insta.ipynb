## Para scrollear todos los seguidores

```python
scroll_box = driver.find_element(By.XPATH,'//div[@class="xyi19xy x1ccrb07 xtf3nb5 x1pc53ja x1lliihq x1iyjqo2 xs83m0k xz65tgg x1rife3k x1n2onr6"]')
last_height, new_height = 0, 1

while new_height != last_height:
    last_height = driver.execute_script("return arguments[0].scrollHeight", scroll_box)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_box)
    time.sleep(3)
    new_height = driver.execute_script("return arguments[0].scrollHeight", scroll_box)
```

## Para sextraer la información

```python
## Funciona con la pantalla maximizada y partida (usuario - link)
wait = WebDriverWait(driver, 10)

# Esperamos hasta que los enlaces de seguidores estén presentes dentro del contenedor de la lista de seguidores
followers = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//a[contains(@href, "/") and not(contains(@href, "/explore/"))]')))

# Creamos una lista para almacenar los usernames y los enlaces a sus perfiles
followers_data = []

# Expresión regular ajustada para capturar usernames válidos (debe comenzar con una letra minúscula)
username_pattern = re.compile(r'^[a-z][a-zA-Z0-9._]+$')

for follower in followers:
    username = follower.text.strip()  # Eliminamos espacios en blanco antes y después
    profile_link = follower.get_attribute('href')  # Extraemos el enlace del perfil
    # Verificamos que el username cumpla con el patrón válido
    if username and username_pattern.match(username):
        followers_data.append({'username': username, 'profile_link': profile_link})

# Mostramos solo los usernames y los enlaces a los perfiles
for follower in followers_data:
    print(f"{follower['username']} - {follower['profile_link']}")
```

## Sacar links de los posts

```python
# Scroll hasta el final de la página
scrolldown = driver.execute_script(
    "window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
match = False
while (match == False):
    last_count = scrolldown
    time.sleep(3)
    scrolldown = driver.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
    if last_count == scrolldown:
        match = True

# Inicializamos una lista vacía para almacenar los enlaces de las publicaciones
posts = []

# Extraemos todos los enlaces de la página
links = driver.find_elements("tag name", "a")
for link in links:
    post_link = link.get_attribute('href')  # Extraemos el atributo href de cada enlace
    if '/p/' in post_link:  # Si el enlace contiene '/p/', es una publicación
        posts.append(post_link)  # Agregamos el enlace a la lista de posts

# Mostramos los enlaces de las publicaciones
for post in posts:
    print(post)
```
